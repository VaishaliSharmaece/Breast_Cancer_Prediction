## 4. REPLACE OUTLIERS WITH NULL
if ratio of mean and st.deviation is greater than 3, I have considered them as outlier and replace it with NAN.

```python
dff=df.select_dtypes(include=['float64'])

dff.mask(dff.sub(dff.mean()).div(dff.std()).abs().gt(3))
```

```python
dff_nan=dff.mask(dff.sub(dff.mean()).div(dff.std()).abs().gt(3))
```

#### count of nan per column

```python
dff_nan.isnull().sum(axis=0) 
```

#### fill nan with mean as all are numerical value and percentage of nan is less then 10% for each column

```python
dff_nan = dff_nan.fillna(dff_nan.mean())
```

```python
dff_nan.isnull().sum(axis=0) 
```

##  5. Calculation of Z Score

 #### while calculating the Z-score we re-scale and center the data and look for data points which are too far from zero. These data points which are way too far from zero will be treated as the outliers. In most of the cases a threshold of 3 or -3 is used i.e if the Z-score value is greater than or less than 3 or -3 respectively, that data point will be identified as outliers.
We will use Z-score function defined in scipy library to detect the outliers.

```python
from scipy import stats
import numpy as np
z = np.abs(stats.zscore(dff_nan))
print(z)
```

```python
threshold = 3
print(np.where(z > 3))
```

## MACHINE LEARNING -CLASSIFICATON PROBLEM PREDICTION MODEL

```python
#8 Classification ML
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import cross_val_score, KFold
from sklearn import metrics
from sklearn.metrics import roc_auc_score
```

```python
df.diagnosis = df.diagnosis.map({'M':0,'B':1})
```

```python
from sklearn.ensemble import RandomForestClassifier

X = df.drop(['diagnosis','id'],axis=1)
y = df[['diagnosis']]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state=42)
model = RandomForestClassifier(n_estimators=100,
                               random_state=101,
                               class_weight='balanced')
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
print('AUC-ROC Score :',roc_auc_score(y_test, y_pred))
print('F1-Score :',f1_score(y_test,y_pred))
print('Report:\n',classification_report(y_test, y_pred))
print('confusion Matrix:\n',confusion_matrix(y_pred,y_test))
```

```python
print('cross validation:',cross_val_score(model, X, y, cv=5).mean())
```

```python
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
```

```python
models = {'RandomForestClassifier':RandomForestClassifier(),
          'LogisitcRegression':LogisticRegression(),
          'GradientBoosingClassifier':GradientBoostingClassifier(),
          'XGBoost':XGBClassifier()}
mn = []
auc = []
f1 = []
acc = []
for model_name, model in models.items():
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    print('\nF1 Score For', model_name,'>>',f1_score(y_test,pred))
    print('AUC-ROC Score :',model_name,'>>',roc_auc_score(y_test, pred))
    print('Report:\n',model_name,'>>',classification_report(y_test, pred))
    print('confusion Matrix:\n',model_name,'>>',confusion_matrix(pred,y_test))
    print('cross validation for',model_name,'::>',cross_val_score(model, X, y, cv=5).mean())
```

## XGBoost Clasifier is the best model to predict the breast cancer as it has the highest accuracy among all the models i.e. 0.98.

### Hyper parameter tuning

```python
from sklearn.model_selection import RandomizedSearchCV,GridSearchCV
models = {'RandomForestClassifier':RandomForestClassifier(),
          'GradientBoosingClassifier':GradientBoostingClassifier(),
          'XGBoost':XGBClassifier()}
params = {
            'max_depth': [30, 40, 50],
            'n_estimators': [20, 30, 40, 50, 100],
}
for model_name, model in models.items():
    grid = GridSearchCV(model, params)
    grid.fit(X_train,y_train)
    y_pred = grid.best_estimator_.predict(X_test)
    print('Scores after HypterParameterTuninig')
    print(model_name,' F1-Score is >> :',f1_score(y_test,y_pred))
```

## after hyperpaarmeter tunning the f1 score of XGBoost classifier is enhanced and is 0.98

###  PREDICT THE TARGET WITH MAXMIUM INDEPENDENT VARIABLES AND FEATURE IMPORTANCE 

```python
from xgboost import plot_importance
import matplotlib.pyplot as plt

# plot feature importance
plot_importance(model)
plt.show()
```

```python
for col,score in zip(X_train.columns,model.feature_importances_):

    print(col,score)
```

```python
from numpy import sort
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import accuracy_score

thresholds = sort(model.feature_importances_)
for thresh in thresholds:
    # select features using threshold
    selection = SelectFromModel(model, threshold=thresh, prefit=True)
    select_X_train = selection.transform(X_train)
    # train model
    selection_model = XGBClassifier()
    selection_model.fit(select_X_train, y_train)
    # eval model
    select_X_test = selection.transform(X_test)
    y_pred = selection_model.predict(select_X_test)
    predictions = [round(value) for value in y_pred]
    accuracy = accuracy_score(y_test, y_pred)
    print("Thresh=%.3f, n=%d, Accuracy: %.2f%%" % (thresh, select_X_train.shape[1], accuracy*100.0))
```
